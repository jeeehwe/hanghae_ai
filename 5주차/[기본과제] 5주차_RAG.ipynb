```python
import bs4
from langchain import hub
from langchain_chroma import Chroma
from langchain_openai import ChatOpenAI
from langchain_openai import OpenAIEmbeddings
from langchain_community.document_loaders import WebBaseLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from google.colab import userdata
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", api_key = userdata.get('OPEN_AI_API_KEY'));
```

```python
import requests
from bs4 import BeautifulSoup, SoupStrainer

url = "https://spartacodingclub.kr/blog/all-in-challenge_winner"
response = requests.get(url)
html = response.text

soup = BeautifulSoup(html, "html.parser", parse_only=SoupStrainer(class_=("editedContent", "css-j3idia")))
texts = soup.get_text()
print(texts)
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=512,
    chunk_overlap=10
)
splits = text_splitter.split_text(texts)
for split in splits:
    print(split)

vectorstore = Chroma.from_texts(
    texts=splits,
    embedding=OpenAIEmbeddings(api_key=userdata.get('OPEN_AI_API_KEY'))
)
```

```python
retriever = vectorstore.as_retriever()

user_msg = "ALL-in 코딩 공모전 수상작들을 요약해줘."
retrieved_docs = retriever.invoke(user_msg)

print(retrieved_docs)

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
prompt = hub.pull("rlm/rag-prompt")

user_prompt = prompt.invoke({"context": format_docs(retrieved_docs), "question": user_msg})
print(user_prompt)
```

### [LOG] 출력 결과

```python
response = llm.invoke(user_prompt)
print(response.content)
```
